# Ufak bir istatistik örneği

İlgili kütüphaneleri import edelim

```python
import numpy as np
import scipy.stats as stats
import pylab as pl
```

## Bir veriye ihtiyacımız var. Bunu bir deneyden toplanmış veriler olarak düşünebiliriz.

```python
h = sorted([186, 176, 158, 180, 186, 168, 168, 164, 178, 170, 189, 195, 172,
     187, 180, 186, 185, 168, 179, 178, 183, 179, 170, 175, 186, 159,
     161, 178, 175, 185, 175, 162, 173, 172, 177, 175, 172, 177, 180])  #sıralanmış
```

## Dağılıma en iyi uyan normal dağılımı bulalım

```python
fit = stats.norm.pdf(h, np.mean(h), np.std(h))  
```

## Grafikleri çizdirelim

```python
pl.plot(h,fit,'-o')

pl.hist(h,normed=True)      #verinin üzerine histogram çiz

pl.show()  
```
